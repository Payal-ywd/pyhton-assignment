```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    total_likes = 0
    
    for like in likes_list:
        total_likes += like
    
    
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 300, 250, 300]
analyze_post_engagement(likes)
```

```python
# Problem 2: Medicine Stock Alert System

def check_medicine_stock(stock):
    
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print("Medicine Stock:", stock)
    print("Status:", status)

# Example
check_medicine_stock(6)
```

```python
# Problem 3: Rainfall Adequacy Checker

def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example
rainfall = [70, 80, 65, 75]
rainfall_checker(rainfall, 70)
```

```python
# Problem 4: Duplicate Account Detection

def detect_duplicates(usernames):
    
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    
    print("Duplicate Accounts Found:", result)


users = ["payal", "rahul", "payal", "amit"]
detect_duplicates(users)
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = []
    
    # Filter prices above ₹2000
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)

# Example
crop_prices = [1500, 2500, 3200, 1800]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: Application Health Checker

def application_health(errors):
    
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)

# Example
application_health(7)
```

```python
# Problem 8: Daily Transaction Limit Checker

def transaction_checker(amount):
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

# Example
transaction_checker(60000)
```

```python
# Problem 9: Attendance Eligibility System

def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    
    # Count attended classes
    for status in attendance_list:
        if status == 1:   # 1 means present
            attended += 1
    
    percentage = (attended / total_classes) * 100
    
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

# Example
attendance = [1,1,1,0,1,1,1,1,0,1]
attendance_checker(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def electricity_bill(units):
    
    # Bill calculation using slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Status:", usage)

# Example
electricity_bill(250)
```
